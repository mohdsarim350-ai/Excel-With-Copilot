#Excel With Copilot

1. Use Copilot to identify and summarize missing values in dataset.

   - Using Copilot, the dataset was analyzed to idnetify missing values. Copilot highlighted missing entries in Region, Category, Units Sold, and Unit Price columns and provided a summary showing the count of missing values in each columns.

  

2. Use Copilot to fill missing values in the Region and Category columns.

   - Copilot was used to automatically fill missing values in the Region and Category columns by analyzing similar rows and patterns in the dataset. After applying the suggestions, all missing text values were successfully filled.

3. Fill missing numeric values in the Units Sold and Unit Price columns.

    - Copilot suggested filling missing numeric values using appropriate methods such as average or median values. After applying the suggestions, the Units Sold and Unit Price columns no longer contained missing values.

4. Create a new calculated column Total Revenue.

   - A new calculated column Total Revenue was created using the formula:

   - Total Revenue = Units Sold x Unit Price.

   - This column calculats the total revenue for each order.

5. Use copilot to find the total price of Order ID = ORDO12.

   - Copilot was used to filter the dataset for Order ID = ORDO123 and calculate the total price using the Total Revenue column. The total price for Order ID ORDO123 was successfully identified.

6. Use Copilot to generate a summary chart of average profit by region.

   - Copilot generated a summary chart showing average profit by region. The chart visually compare profitability across regions and helps identify the best-performing region.

7. Create a pivot table summarizing total sales by category and region.

   - A PivotTable was created with:

   - Rows: Category

   - Columns: Region

   - Values: Sum of Sales

   - The PivotTable summarizes total sales by category and region.

8. Build a dashboard combining key KPIs and charts.

   - A dashboard was created that includes:

   - Total Sales KPI

   - Total Revenue KPI

   - Average Profit

   - Sales by Region chart

   - Category-wise sales chart

   - The dashboard provides a quick overview of business performance.

9. Use Copilot to generate a summary report.

   - Copilot was used to generate a summary report highlighting:

   - Overall sales performance

   - Regional and category trends

   - Key insights and observations

   - The report provides a concise summary of the dataset and analysis.
